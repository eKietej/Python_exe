### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

### Step 2. Create the DataFrame that should look like the one below.

```python
data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'age': [42, 52, 36, 24, 73],
    'female': [0, 1, 1, 0, 1],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}

df = pd.DataFrame(data)
df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.figure(figsize=(8,6))
plt.scatter(df['preTestScore'], df['postTestScore'], 
            s=df['age']*3,
            alpha=0.6, 
            c='skyblue', 
            edgecolors='w')
for i in range(len(df)):
    plt.text(df['preTestScore'][i]+0.5, df['postTestScore'][i]+0.5, 
             df['first_name'][i], fontsize=9)
plt.xlabel('Pre Test Score')
plt.ylabel('Post Test Score')
plt.title('Scatterplot of Pre vs Post Test Scores (size by age)')

plt.show()
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python


plt.figure(figsize=(8,6))
colors = df['female'].map({0:'blue', 1:'pink'})
plt.scatter(df['preTestScore'], df['postTestScore'],
            s=df['postTestScore']*4.5,
            c=colors,
            alpha=0.6,
            edgecolors='w')
for i in range(len(df)):
    plt.text(df['preTestScore'][i]+0.5, df['postTestScore'][i]+0.5, 
             df['first_name'][i], fontsize=9)
male_patch = mpatches.Patch(color='blue', label='Male')
female_patch = mpatches.Patch(color='pink', label='Female')
plt.legend(handles=[male_patch, female_patch])
plt.xlabel('Pre Test Score')
plt.ylabel('Post Test Score')
plt.title('Scatterplot of Pre vs Post Test Scores (size by postTestScore, color by sex)')

plt.show()
```
