### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

```python
url = "https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv"
```

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv(url, sep = ",")
tips
```

### Step 4. Delete the Unnamed 0 column

```python
if 'Unnamed: 0' in tips.columns:
    tips = tips.drop(columns=['Unnamed: 0'])
tips
```

### Step 5. Plot the total_bill column histogram

```python
plt.figure(figsize=(8,6))
plt.hist(tips['total_bill'], bins=20, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Histogram of Total Bill')
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.figure(figsize=(8,6))
plt.scatter(tips['total_bill'], tips['tip'], 
            color='orange', 
            alpha=0.6, 
            edgecolors='w', 
            s=50)

plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
plt.figure(figsize=(8,6))
colors = tips['sex'].map({'Male':'blue', 'Female':'red'})
plt.scatter(tips['total_bill'], tips['tip'],
            s=tips['size']*50,
            c=colors,
            alpha=0.6,
            edgecolors='w')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot of Total Bill vs Tip (size by group size, color by sex)')
import matplotlib.patches as mpatches
male_patch = mpatches.Patch(color='blue', label='Male')
female_patch = mpatches.Patch(color='red', label='Female')
plt.legend(handles=[male_patch, female_patch])
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Distribution of Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='tip', y='day', hue='sex', data=tips, s=100, palette={'Male':'blue','Female':'red'})
plt.xlabel('Tip')
plt.ylabel('Day')
plt.title('Scatter Plot of Tip vs Day (colored by sex)')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', hue='time', data=tips, palette='Set2')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Total Bill per Day by Time (Lunch vs Dinner)')
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
tips_lunch = tips[tips['time']=='Lunch']
tips_dinner = tips[tips['time']=='Dinner']
fig, axes = plt.subplots(1, 2, figsize=(12,5), sharey=True)
axes[0].hist(tips_lunch['tip'], bins=15, color='blue', edgecolor='black', alpha=0.7)
axes[0].set_title('Tip Distribution - Lunch')
axes[0].set_xlabel('Tip')
axes[0].set_ylabel('Frequency')
axes[1].hist(tips_dinner['tip'], bins=15, color='red', edgecolor='black', alpha=0.7)
axes[1].set_title('Tip Distribution - Dinner')
axes[1].set_xlabel('Tip')
plt.tight_layout()
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
tips_male = tips[tips['sex']=='Male']
tips_female = tips[tips['sex']=='Female']
fig, axes = plt.subplots(1, 2, figsize=(14,6), sharey=True)
colors_male = tips_male['smoker'].map({'Yes':'red','No':'green'})
axes[0].scatter(tips_male['total_bill'], tips_male['tip'], c=colors_male, s=80, alpha=0.6, edgecolors='w')
axes[0].set_title('Male')
axes[0].set_xlabel('Total Bill')
axes[0].set_ylabel('Tip')
import matplotlib.patches as mpatches
male_smoker = mpatches.Patch(color='red', label='Smoker')
male_nonsmoker = mpatches.Patch(color='green', label='Non-Smoker')
axes[0].legend(handles=[male_smoker, male_nonsmoker])
colors_female = tips_female['smoker'].map({'Yes':'red','No':'green'})
axes[1].scatter(tips_female['total_bill'], tips_female['tip'], c=colors_female, s=80, alpha=0.6, edgecolors='w')
axes[1].set_title('Female')
axes[1].set_xlabel('Total Bill')
female_smoker = mpatches.Patch(color='red', label='Smoker')
female_nonsmoker = mpatches.Patch(color='green', label='Non-Smoker')
axes[1].legend(handles=[female_smoker, female_nonsmoker])
plt.suptitle('Total Bill vs Tip by Sex and Smoker Status', fontsize=16)
plt.tight_layout()
plt.show()
```
